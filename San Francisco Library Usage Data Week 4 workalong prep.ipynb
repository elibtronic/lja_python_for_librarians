```python
import pandas as pd

dataframe = pd.read_csv("Library_Usage.csv",skipinitialspace = True)
dataframe.dropna()
dataframe.head(5)
```

```python
#drop columns we don't need
dataframe.pop("Age Range")
dataframe.pop("Outside of County")
dataframe.pop("Home Library Code")
dataframe.pop("Year Patron Registered")
dataframe.pop("Circulation Active Month")
dataframe.pop("Notice Preference Code")
dataframe.pop("Provided Email Address")
dataframe.pop("Notice Preference Definition")
dataframe.pop("Supervisor District")
dataframe.pop("Home Library Definition")
dataframe.pop("Patron Type Code")

dataframe
```

```python
dataframe["Patron Type Definition"].unique()
```

```python
#None?
dataframe["Circulation Active Year"].replace({'None':0},inplace=True)
dataframe["Circulation Active Year"].unique()
```

```python
# Get the categories

dataframe["Patron Type Definition"].replace({\
                            'SENIOR':1, 
                            'ADULT':1, 
                            'DIGITAL ACCESS CARD':0, 
                            'RETIRED STAFF':0, 
                            'STAFF':0,
                            'SPECIAL':0,
                            'WELCOME':0,
                            'TEACHER CARD':0,
                            'AT USER ADULT':0,
                            'VISITOR':0,
                            'BOOKS BY MAIL':0, 
                            'JUVENILE':2,
                            'AT USER SENIOR':0,
                            'FRIENDS FOR LIFE':0,
                            'AT USER WELCOME':0,
                            'AT USER TEEN':0,
                            'YOUNG ADULT':2,
                            'AT USER JUVENILE':0
                            },inplace=True)

```
